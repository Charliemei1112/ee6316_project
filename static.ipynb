```python
import pandas as pd
import numpy as np

# read file "static.csv"
df = pd.read_csv("static.csv")
df_09 = pd.read_csv("static_09.csv")
df_11 = pd.read_csv("static_11.csv")

# print the first 5 rows of the dataframe
print(df.head())

# print column names
print(df.columns)
```

```python
# drop column "/Vout- X"
df = df.drop(columns=["/Vout- X"])
df_09 = df_09.drop(columns=["/Vout- X"])
df_11 = df_11.drop(columns=["/Vout- X"])

# change column name "Vout+ X" to "code"
df = df.rename(columns={"/Vout+ X": "code"})
df_09 = df_09.rename(columns={"/Vout+ X": "code"})
df_11 = df_11.rename(columns={"/Vout+ X": "code"})

# add column Vdiff
df["Vdiff"] = df["/Vout- Y"] - df["/Vout+ Y"]
df_09["Vdiff"] = df_09["/Vout- Y"] - df_09["/Vout+ Y"]
df_11["Vdiff"] = df_11["/Vout- Y"] - df_11["/Vout+ Y"]

# drop column "/Vout+ Y"
df = df.drop(columns=["/Vout+ Y"])
df_09 = df_09.drop(columns=["/Vout+ Y"])
df_11 = df_11.drop(columns=["/Vout+ Y"])

# drop column "/Vout- Y"
df = df.drop(columns=["/Vout- Y"])
df_09 = df_09.drop(columns=["/Vout- Y"])
df_11 = df_11.drop(columns=["/Vout- Y"])

print(df.head())
print(df_09.head())
print(df_11.head())
```

```python
print(df.shape)
LSB = (df["Vdiff"].max() - df["Vdiff"].min()) / 255
LSB_09 = (df_09["Vdiff"].max() - df_09["Vdiff"].min()) / 255
LSB_11 = (df_11["Vdiff"].max() - df_11["Vdiff"].min()) / 255
```

```python
# compute DNL for each code
DNL = []
DNL_09 = [] 
DNL_11 = []

for i in range (0, 255):
    DNL.append((df["Vdiff"][i+1] - df["Vdiff"][i])/LSB - 1)
    DNL_09.append((df_09["Vdiff"][i+1] - df_09["Vdiff"][i])/LSB_09 - 1)
    DNL_11.append((df_11["Vdiff"][i+1] - df_11["Vdiff"][i])/LSB_11 - 1)

print(DNL)

# check if there are DNL < -1 or DNL > 1
for dnl in DNL:
    if dnl < -1 or dnl > 1:
        print("DNL error")
        break

for dnl in DNL_09:
    if dnl < -1 or dnl > 1:
        print("DNL_09 error")
        break

for dnl in DNL_11:
    if dnl < -1 or dnl > 1:
        print("DNL_11 error")
        break

# find min, max, and average, std of DNL
print(f"min(DNL): {min(DNL)}")
print(f"max(DNL): {max(DNL)}")
print(f"mean(DNL): {np.mean(DNL)}")
print(f"std(DNL): {np.std(DNL)}")

print(f"min(DNL_09): {min(DNL_09)}")
print(f"max(DNL_09): {max(DNL_09)}")
print(f"mean(DNL_09): {np.mean(DNL_09)}")
print(f"std(DNL_09): {np.std(DNL_09)}")

print(f"min(DNL_11): {min(DNL_11)}")
print(f"max(DNL_11): {max(DNL_11)}")
print(f"mean(DNL_11): {np.mean(DNL_11)}")
print(f"std(DNL_11): {np.std(DNL_11)}")

INL = []
INL_09 = []
INL_11 = []

for i in range (0, 254):
    INL.append(np.sum(DNL[:i+1]))
    INL_09.append(np.sum(DNL_09[:i+1]))
    INL_11.append(np.sum(DNL_11[:i+1]))

print(INL)
print(INL_09)
print(INL_11)

# check if there are INL < -2 or INL > 2:
for inl in INL:
    if inl < -2 or inl > 2:
        print("INL error")
        break

for inl in INL_09:
    if inl < -2 or inl > 2:
        print("INL_09 error")
        break

for inl in INL_11:
    if inl < -2 or inl > 2:
        print("INL_11 error")

# print min, max, mean, std of INL
print(f"min(INL): {min(INL)}")
print(f"max(INL): {max(INL)}")
print(f"mean(INL): {np.mean(INL)}")
print(f"std(INL): {np.std(INL)}")

print(f"min(INL_09): {min(INL_09)}")
print(f"max(INL_09): {max(INL_09)}")
print(f"mean(INL_09): {np.mean(INL_09)}")
print(f"std(INL_09): {np.std(INL_09)}")

print(f"min(INL_11): {min(INL_11)}")
print(f"max(INL_11): {max(INL_11)}")
print(f"mean(INL_11): {np.mean(INL_11)}")
print(f"std(INL_11): {np.std(INL_11)}")
```
