```python
import pandas as pd
import numpy as np

# read file "static.csv"
df = pd.read_csv("DAC_static.csv")
df_09 = pd.read_csv("DAC_static_09.csv")
df_11 = pd.read_csv("DAC_static_11.csv")

# print the first 5 rows of the dataframe
print(df.head())

# print column names
print(df.columns)
```

```python
# drop column "/Vout- X"
df = df.drop(columns=["/Vout- X"])
df_09 = df_09.drop(columns=["/Vout- X"])
df_11 = df_11.drop(columns=["/Vout- X"])

# change column name "Vout+ X" to "code"
df = df.rename(columns={"/Vout+ X": "code"})
df_09 = df_09.rename(columns={"/Vout+ X": "code"})
df_11 = df_11.rename(columns={"/Vout+ X": "code"})

# add column Vdiff
df["Vdiff"] = df["/Vout+ Y"] - df["/Vout- Y"]
df_09["Vdiff"] = df_09["/Vout+ Y"] - df_09["/Vout- Y"]
df_11["Vdiff"] = df_11["/Vout+ Y"] - df_11["/Vout- Y"]

# drop column "/Vout+ Y"
df = df.drop(columns=["/Vout+ Y"])
df_09 = df_09.drop(columns=["/Vout+ Y"])
df_11 = df_11.drop(columns=["/Vout+ Y"])

# drop column "/Vout- Y"
df = df.drop(columns=["/Vout- Y"])
df_09 = df_09.drop(columns=["/Vout- Y"])
df_11 = df_11.drop(columns=["/Vout- Y"])

print(df.head())
print(df_09.head())
print(df_11.head())
```

```python
print(df.shape)
LSB = (df["Vdiff"].max() - df["Vdiff"].min()) / 255
LSB_09 = (df_09["Vdiff"].max() - df_09["Vdiff"].min()) / 255
LSB_11 = (df_11["Vdiff"].max() - df_11["Vdiff"].min()) / 255

print(f"LSB: {LSB}")
print(f"LSB_09: {LSB_09}")
print(f"LSB_11: {LSB_11}")
```

```python
# compute DNL for each code
DNL = []
DNL_09 = [] 
DNL_11 = []

for i in range (0, 255):
    DNL.append((df["Vdiff"][i+1] - df["Vdiff"][i])/LSB - 1)
    DNL_09.append((df_09["Vdiff"][i+1] - df_09["Vdiff"][i])/LSB_09 - 1)
    DNL_11.append((df_11["Vdiff"][i+1] - df_11["Vdiff"][i])/LSB_11 - 1)

# print(DNL)

# check if there are DNL < -1 or DNL > 1
for dnl in DNL:
    if dnl < -1 or dnl > 1:
        print("DNL error, did not meet the design requirement")
        break

for dnl in DNL_09:
    if dnl < -1 or dnl > 1:
        print("DNL_09 error, did not meet the design requirement")
        break

for dnl in DNL_11:
    if dnl < -1 or dnl > 1:
        print("DNL_11 error, did not meet the design requirement")
        break

# find min, max, and average, std of DNL,
print("DNL analysis:")
print(f"min: {min(DNL):.3f}, max: {max(DNL):.3f}, mean: {np.mean(DNL):.3f}, std: {np.std(DNL):.3f}")

print("DNL_09 analysis:")
print(f"min: {min(DNL_09):.3f}, max: {max(DNL_09):.3f}, mean: {np.mean(DNL_09):.3f}, std: {np.std(DNL_09):.3f}")

print("DNL_11 analysis:")
print(f"min: {min(DNL_11):.3f}, max: {max(DNL_11):.3f}, mean: {np.mean(DNL_11):.3f}, std: {np.std(DNL_11):.3f}")

INL = []
INL_09 = []
INL_11 = []

for i in range (0, 255):
    INL.append(np.sum(DNL[:i+1]))
    INL_09.append(np.sum(DNL_09[:i+1]))
    INL_11.append(np.sum(DNL_11[:i+1]))

# check if there are INL < -2 or INL > 2:
for inl in INL:
    if inl < -2 or inl > 2:
        print("INL error, did not meet the design requirement")
        break

for inl in INL_09:
    if inl < -2 or inl > 2:
        print("INL_09 error, did not meet the design requirement")
        break

for inl in INL_11:
    if inl < -2 or inl > 2:
        print("INL_11 error, did not meet the design requirement")

# print min, max, mean, std of INL
print("INL analysis:")
print(f"min: {min(INL):.3f}, max: {max(INL):.3f}, mean: {np.mean(INL):.3f}, std: {np.std(INL):.3f}")

print("INL_09 analysis:")
print(f"min: {min(INL_09):.3f}, max: {max(INL_09):.3f}, mean: {np.mean(INL_09):.3f}, std: {np.std(INL_09):.3f}")

print("INL_11 analysis:")
print(f"min: {min(INL_11):.3f}, max: {max(INL_11):.3f}, mean: {np.mean(INL_11):.3f}, std: {np.std(INL_11):.3f}")

# find Differential Peak-to-Peak Voltage
Vdiff_pp = df["Vdiff"].max() - df["Vdiff"].min()
Vdiff_pp_09 = df_09["Vdiff"].max() - df_09["Vdiff"].min()
Vdiff_pp_11 = df_11["Vdiff"].max() - df_11["Vdiff"].min()

print(f"Vdiff_pp: {Vdiff_pp:.3f}")
print(f"Vdiff_pp_09: {Vdiff_pp_09:.3f}")
print(f"Vdiff_pp_11: {Vdiff_pp_11:.3f}")
```
